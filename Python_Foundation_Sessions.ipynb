## S4 · Concept 2: Default Params & Keyword Args
### 📖 STUDY

```python
# Default parameter — used when caller doesn't provide it
def calculate_fee(amount, rate=0.01):
    return amount * rate

print(calculate_fee(10000))         # 100.0 (uses default)
print(calculate_fee(10000, 0.02))   # 200.0 (overrides)

# Keyword arguments — pass by name, order doesn't matter
def trade_summary(ticker, action, qty):
    return f"{action} {qty} shares of {ticker}"

print(trade_summary(qty=50, ticker="AAPL", action="BUY"))
```

### 🙈 HIDE
You created a function with a default parameter (rate=0.01) that gets used when the caller skips it, but can be overridden. Then you wrote a function and called it using keyword arguments in a shuffled order to show that named args don't need to be in order.

### ✍️ WRITE — Function with a default param (call with and without it), function called with keyword args out of order

```python
# YOUR CODE HERE

def gst_cal(amount,gst=0.18,cess=0.0):
    gst_charged= amount * (gst if gst < 1.0 else gst/100)
    cess_charged= amount * (cess if cess < 1.0 else cess/100)
    final_amt= amount + gst_charged + cess_charged
    return f"\n Amount: {amount} \n GST: {gst_charged} \n CESS: {cess_charged} \n Total: {final_amt} "


print(gst_cal(500))
print(gst_cal(500, gst=10, cess=1))
```

## S4 · Concept 3: Lambda & max(key=)
### 📖 STUDY

```python
# Lambda = a tiny one-line function (no name needed)
double = lambda x: x * 2
print(double(5))   # 10

# Real power: using lambda as a key function
stocks = [
    {"ticker": "AAPL", "price": 189.50},
    {"ticker": "GOOG", "price": 178.30},
    {"ticker": "MSFT", "price": 412.00},
]

# max() with key= tells Python WHAT to compare by
top = max(stocks, key=lambda s: s["price"])
print(top["ticker"], top["price"])   # MSFT 412.0
```

### 🙈 HIDE
You created a lambda — a one-line anonymous function. Then you had a list of stock dicts and used max() with key=lambda to find the stock with the highest price. The lambda told max() to compare each dict by its 'price' value.

### ✍️ WRITE — Create a lambda, then use max(key=lambda) on a list of dicts to find the highest-priced stock

```python
# YOUR CODE HERE
triple= lambda x: x * 3
print(triple(5))


store_product = [
    {"Product": "Parle", "Price": 10},
    {"Product": "kellogs", "Price": 350},
    {"Product": "Britania", "Price": 50},
    {"Product": "UniBicks", "Price": 100}
]

high_price= max(store_product, key= lambda x: x["Price"])

print(high_price["Product"], high_price["Price"])
```

## S5 · Concept 2: Specific Error Types
### 📖 STUDY

```python
data = {"ticker": "AAPL"}
prices = [100, 200, 300]

# Catch SPECIFIC errors — much better than bare except
try:
    print(data["price"])       # KeyError
except KeyError:
    print("Key not found in dict")

try:
    print(prices[10])          # IndexError
except IndexError:
    print("Index out of range")

try:
    result = "10" + 5          # TypeError
except TypeError as e:
    print(f"Type mismatch: {e}")
```

## S5 · Concept 3: Defensive Coding
### 📖 STUDY

```python
def safe_divide(a, b):
    # Guard clause — check BEFORE doing the risky thing
    if b == 0:
        return "Cannot divide by zero"
    return a / b

print(safe_divide(100, 4))    # 25.0
print(safe_divide(100, 0))    # Cannot divide by zero

# .get() is defensive coding for dicts
stock = {"ticker": "AAPL"}
price = stock.get("price", "N/A")
print(f"Price: {price}")   # Price: N/A
```

### 🙈 HIDE

## S6 · Concept 3: Methods
### 📖 STUDY

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = []

    def add_stock(self, ticker, shares, price):
        self.holdings.append({"ticker": ticker, "shares": shares, "price": price})

    def total_value(self):
        total = 0
        for h in self.holdings:
            total += h["shares"] * h["price"]
        return total

    def summary(self):
        return f"{self.owner}'s portfolio: ${self.total_value():,.2f}"

p = Portfolio("Sachin")
p.add_stock("AAPL", 50, 189.50)
p.add_stock("GOOG", 20, 178.30)
print(p.summary())   # Sachin's portfolio: $13,041.00

print(p.total_value())   # Sachin's portfolio: $13,041.00
```

### 🙈 HIDE
You built a full Portfolio class with three methods: add_stock stores holdings, total_value loops through and sums shares*price, and summary calls total_value inside an f-string. Methods can call other methods on the same object using self.

### ✍️ WRITE — Full Portfolio class: add_stock (with price), total_value method, summary method that calls total_value

```python
# YOUR CODE HERE
class Portfolio:
    def __init__(self, owner):
        self.owner= owner
        self.holdings= []

    def add_stocks(self, ticker, quantity, price):
        self.holdings.append({"ticker": ticker, "quantity": quantity, "price": price})

    def total_value(self):
        total = 0
        for h in self.holdings:
            total += h["quantity"] * h["price"]
        return total
    
    def summary(self):
        return f"Portfolio Owner: {self.owner} has total stock value of ${self.total_value():,.0f}"

    def all_stocks(self):
        output= ""
        for h in self.holdings:
            output +=f" Stock Name: {h['ticker']} & Total Amount Invested: ${h['quantity'] * h['price']:,.2f} \n"
        return output
    
            

p= Portfolio("Sachin")

p.add_stocks("AAPL", 10, 100)
p.add_stocks("NVDI", 30, 800)

print(p.all_stocks())

print(p.summary())

```
