## Challenge 3: Bulletproof Calculator
**Concepts tested:** functions, try/except, specific error types, guard clauses, default params

Write a function `safe_calc(a, b, operation="add")` that:
- Supports 4 operations: `"add"`, `"sub"`, `"mul"`, `"div"`
- Has a guard clause: if `operation` is not one of the 4, return `"Unknown operation"`
- For division: has a guard clause for zero
- Wraps the whole thing in try/except to catch `TypeError` (in case someone passes a string as `a` or `b`)

Test it with these calls:
```python
print(safe_calc(10, 5))              # 15
print(safe_calc(10, 5, "div"))       # 2.0
print(safe_calc(10, 0, "div"))       # Cannot divide by zero
print(safe_calc(10, 5, "power"))     # Unknown operation
print(safe_calc("ten", 5))           # TypeError message
```

````python
# YOUR CODE HERE
def safe_calc(a, b, operation="add"):
    try:
        if operation=="add":
            return a+b
        elif operation=="sub":
            return a-b
        elif operation=="mul":
            return a*b
        elif operation=="div":
            if b==0:
                return "Cannot divide by zero"
            else:
                return a/b
        else:
            return "Unknown Operation"
    except TypeError:
        return "Incorrect data type"


print(safe_calc(10, 5))              # 15
print(safe_calc(10, 5, "div"))       # 2.0
print(safe_calc(10, 0, "div"))       # Cannot divide by zero
print(safe_calc(10, 5, "power"))     # Unknown operation
print(safe_calc("ten", 5))           # TypeError message
````

## Challenge 5: The Full Class
**Concepts tested:** classes, __init__, self., methods, error handling, f-strings, loops

Build a class called `TradingAccount` from scratch:

**__init__** takes `owner` and `balance` (starting cash)
- Store both as instance attributes
- Also create an empty list `self.trades`

**buy(ticker, shares, price)**
- Calculate cost = shares * price
- If cost > self.balance, return `"Insufficient funds"`
- Otherwise: subtract cost from balance, append trade dict to self.trades, return `"Trade executed"`

**sell(ticker, shares, price)**
- Add (shares * price) to balance
- Append trade dict to self.trades (action = "SELL")

**summary()**
- Return: `"Owner: Sachin | Balance: $XX,XXX.XX | Trades: N"`

Test:
```python
acc = TradingAccount("Sachin", 50000)
print(acc.buy("AAPL", 100, 189.50))   # Trade executed
print(acc.buy("MSFT", 500, 412.00))   # Insufficient funds
acc.sell("AAPL", 50, 195.00)
print(acc.summary())
```

````python
# YOUR CODE HERE
class TradingAccount:
    def __init__(self, owner, balance):
        self.owner= owner
        self.balance= balance
        self.trades = []

    def buy(self, ticker, shares, price):
        self.shares=shares
        self.price=price
        self.cost= self.shares * self.price
        if self.cost > self.balance:
            return f"Insufficient funds,Trade Cost: ${self.cost:,.2f}, Current Account Balance: ${self.balance:,.2f}"
        else:
            trd_ext_log_b = {"ticker": ticker, "shares": shares,"price": price, "action": "BUY"}
            self.trades.append(trd_ext_log_b)
            self.balance -= self.cost
            return f"Trade executed, Trade Cost: ${self.cost:,.2f}, Current Account Balance: ${self.balance:,.2f}"

    def sell(self, ticker, shares, price):
        self.shares=shares
        self.price=price
        sold_amt= self.shares * self.price
        self.balance += sold_amt
        trd_ext_log_s = {"ticker": ticker, "shares": shares,"price": price, "action": "SELL"}
        self.trades.append(trd_ext_log_s)
        return f"Trade executed, Shares sold for: ${sold_amt:,.2f}, Current Account Balance: ${self.balance:,.2f}"

    def summary(self):
        return f"Owner: {self.owner} | Balance: ${self.balance:,.2f} | Trades: {len(self.trades)}"

    def show_trades(self):
        output = ""
        for k in self.trades:
            output += f"{k} \n"
        return f"------Trade Logs------ \n{output}"


acc = TradingAccount("Sachin", 50000)

print(acc.buy("AAPL", 100, 189.50))   # Trade executed

print(acc.buy("MSFT", 500, 412.00))   # Insufficient funds

print(acc.sell("AAPL", 50, 195.00))

print(acc.summary())

print(acc.show_trades())
````
